# ***Numpy indexing and selection***

```python
larr = np.arange(1 , 11)
# grab the index value and slice the array
print(larr[8])
print(larr[0 : 5])
print(larr[4 : ]) # here its print 4 to all the way end here actually we pass the index not the value its grab the value after that index here
```

```python
randomarray = np.random.randint(0 , 11 , 10)
print(randomarray)
# in this randomarray i set the first 4 number is 7 by
randomarray[0 : 4] = 7
print(randomarray)
# dont get confused eveytime the number is changing
```

**Numpy dont set copies of array instant its change the array if you define the array variable name same for other array variable name**

```python
arr = np.arange(0 , 11)
print(arr)
```

```python
changeArray = arr[:]
changeArray[:] = 99
print(changeArray)
print(arr)
# here updated array and old arry two change
```

```python
# if you want to change the value but not affect the old array then use copy
new_Arr = np.random.randint(0 , 20 , 10)
print(new_Arr)
new_Arr_copy = new_Arr.copy()
new_Arr_copy[:5] = np.arange(0 , 5)
print(new_Arr_copy)
```

```python
# using single bracket comma notation
new_2D = np.random.randint(1 , 100 ,size=(3,3))
print(new_2D)

# here i want to grab the 2nd row 3rd item
print(new_2D[1 , 2])
```

```python
# here i grab the 2nd and 3rd row 2nd and 3rd coloum
new_2D_array = np.random.randint(0 , 10 ,  size=(3 , 5))
print(new_2D_array)
# now grab
print(new_2D_array[1 : 3 , 1 : 3])
```

```python
full_Array = np.arange(1 , 51).reshape(5 ,10)
print(full_Array)
# grabbing 12 to 30 in array format
print("# grabbing 12 to 30 in array format")
print(full_Array[1 : 3 , 1 : ])
```

# **comparison operators python**

```python
def fun():
  random_Newarray = np.random.randint(0 , 20 , size=[1 , 12])
  boolarray = random_Newarray > 5
  print(f"First array {random_Newarray}")
  print(f"2nd array True / false the number greater than 5 {boolarray}")
  print(f"finally the number itself that greater than 5 {random_Newarray[boolarray]}")

fun()
```
